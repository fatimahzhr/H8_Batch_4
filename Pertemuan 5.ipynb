# Combining Multiple Datasets

```python
city_data
```

```python
another_city_data = pd.DataFrame(
    {
    "revenue": [4500, 8000],
    "employee_count": [2,2]
    }
    ,index=["New York", "Barcelona"]
)
```

```python
another_city_data
```

```python
all_city_data = pd.concat([city_data,another_city_data])
```

```python
all_city_data
```

```python
city_countries = pd.DataFrame({
    "country": ["Holland", "Japan", "Holland", "Canada", "Spain"],
    "capital": [1,1,0,0,0]},
    index=["Amsterdam", "Tokyo", "Rotterdam", "Toronto", "Barcelona"]
)
```

```python
city_countries
```

```python
cities = pd.concat([all_city_data, city_countries], axis=1, sort=False)
```

```python
cities
```

```python
pd.concat([all_city_data, city_countries], axis=1, join="inner")
```

```python
all_city_data.merge(city_countries, left_index=True, right_index=True)
```

```python
countries = pd.DataFrame({
    "population": [17,127,37],
    "continent": ["Europe", "Asia", "North America"]},
    index=["Holland", "Japan", "Canada"])
```

```python
countries
```

```python
cities
```

```python
pd.merge(left=cities, right=countries, left_on="country", right_index=True)
```

```python
cities.merge(right=countries, left_on="country", right_index=True)
```

```python
pd.merge(
    cities,
    countries,
    left_on="country",
    right_index=True,
    how="left")
```
