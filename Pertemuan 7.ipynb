- ### Line Plots

```python
# Plot a line graph of immigration from Haiti
haiti = df_can.loc['Haiti', years]
haiti.head(7)
```

```python
haiti.plot()
```

```python
haiti.index = haiti.index.map(int)
haiti.plot(kind='line')
plt.title('Immigration from Haiti')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')

plt.show()
```

```python
haiti.plot(kind='line')
plt.title('Immigration from Haiti')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')

# annotate the 2010 earthquake
# syntax: plt.text(x,y,label)
plt.text(2000,6000, '2010 Earthquake')

plt.show()
```

```python
# Compare the number of immigrants fron India and China fr 1980-2013
China_India = df_can.loc[['China', 'India'], years]
China_India
```

```python
China_India.plot(kind='line')
```

```python
df_CI = China_India.transpose()
df_CI.head()
```

```python
df_CI.plot(kind='line')
plt.title('Immigration from China and India')
plt.ylabel('Number of Immigrants')
plt.xlabel('Year')
plt.show()
```

```python
# Compare the trend of top 5 countries that contributed the most to immigration to canada

df_can.sort_values(by='Total', ascending=False, axis=0, inplace=True)
df_top5 = df_can.head(5)
df_top5 = df_top5[years].transpose()
df_top5.index = df_top5.index.map(int)
df_top5.plot(kind='line', figsize=(14,8))

plt.title('Immigration Trend of Top 5 Countries')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')

plt.show()
```

- ### Area Plots

```python
df_can.sort_values(['Total'], ascending=False, axis=0, inplace=True)
df_top5 = df_can.head()
df_top5 = df_top5[years].transpose()
df_top5.head()
```

```python
df_top5.Index = df_top5.index.map(int)
df_top5.plot(kind = 'area', stacked=False, figsize=(20,10))

plt.title('Immigration Trend of Top 5 Countries')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')

plt.show()
```

```python
# Scripting Layer
df_top5.plot(kind='area', alpha=0.25, stacked=False, figsize=(20,15))

plt.title('Immigration Trend of Top 5 Countries')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')

plt.show()
```

```python
# Artist Layer
ax = df_top5.plot(kind='area', alpha=0.35, figsize=(20,10))

ax.set_title('Immigration Trend of Top 5 Countries')
ax.set_ylabel('Number of Immigrants')
ax.set_xlabel('years')
```

```python
top_15 = df_can.sort_values(['Total'], ascending=False, axis=0).head(15)
top_15
```

```python
years_80s = range(1980,1990)
years_90s = range(1990,2000)
years_00s = range(2000, 2010)

df_80s = top_15.loc[:, years_80s].sum(axis=1)
df_90s = top_15.loc[:, years_90s].sum(axis=1)
df_00s = top_15.loc[:, years_00s].sum(axis=1)

new_df = pd.DataFrame({'1980s': df_80s, '1990s': df_90s, '2000s': df_00s})
new_df.head()
```

```python
new_df.describe()
```
